3. Tratamento de valores duplicados

```python
net.drop_duplicates(inplace=True)
net.head()
```

4. Preparacao de variaveis

```python
net.drop(['date_added', 'rating'], axis=1, inplace=True)
```

```python
net.rename(columns={'type':'Type','title': 'Title', 'director':'Director', 'country':'Country','release_year':'Year','duration':'Duration', 'listed_in':'Category'}, inplace = True)
net.head()
```

```python
net['Year'].agg(['median','mean'])
```

```python
net['Type'].value_counts()
```

```python
net['Country'].value_counts()
```

```python
net_gp_type_dur = net.groupby(['Type', 'Duration'])
net_gp_type_dur.first()
```

```python
net.value_counts('Category')
```

```python
net_gp_cat_y = net.groupby(['Category', 'Year'])
net_gp_cat_y.first()
```

```python
net.loc[net['Type'] == 'Movie', 'Duration'] = net[net['Type'] == 'Movie']['Duration'].str.replace(' min', '').astype(int)
```

```python
net.loc[net['Type'] == 'TV Show', 'Duration'] = net[net['Type'] == 'TV Show']['Duration'].str.replace(' Seasons?', '', regex=True).astype(int)
```

```python
net['Duration'].head()
```

```python
net_action_90 = net[(net['Category'] == 'Action & Adventure') & (net['Duration'] > 90)]
net_action_90
```

```python
net['Decade']=(net['Year'] // 10) * 10
net.head()
```

5. Exportacao de dados limpos

```python
net.to_csv('cleaned_netflix1.csv')
```

Visualizacao dos dados

```python
sns.pairplot(net)
```

```python
sns.countplot(data=net , x="Type")
```

```python
plt.figure(figsize = (15,7))
top_5 = net['Director'].value_counts().drop('Not Given').head(5)
sns.barplot(x = top_5.values, y  = top_5.index , edgecolor = 'k', linewidth = 1, saturation = 1, color = 'green', alpha =0.8)

plt.title("Top 5 Directors", fontsize = 14, fontweight = 'heavy')
plt.show()
```

```python
plt.figure(figsize = (20,8))
dark_palette = sns.color_palette("dark:#2ecc71", n_colors=3)
net_c = net['Category'].value_counts().head(10)
count = sns.barplot(x = net_c.index, y=net_c.values , alpha = 0.9, saturation = 1  , edgecolor = 'k', linewidth = 2, palette = dark_palette)
for p in count.patches:
    per = round( (p.get_height()/net.shape[0])*100 ,1 )
    x = p.get_x() 
    y = p.get_height() +7
    count.annotate('{0}%'.format(per), (x, y ), fontsize = 7)
    
plt.xticks(rotation = 90)
plt.title('Top 10 Genres  In Netflix', fontsize = 14, fontweight = 'heavy')
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 
country_counts = net[net['Type'] == 'TV Show']['Country'].value_counts().head(15)

country_counts.plot(kind='bar', color='skyblue')

plt.xlabel('Country')
plt.ylabel('Number of TV Shows')
plt.title('Number of Netflix TV Shows Produced by Country (Top 15)')
plt.xticks(rotation=45)  
plt.tight_layout() 
plt.show()

```

```python
plt.figure(figsize=(12, 6))

g = sns.catplot(data=net[net['Type'] == 'Movie'], x='Decade', y='Duration', hue='Type', kind='bar')

g.set_axis_labels("Decade", "Duration")
g.set_titles("Relationship between Movie Duration and Decade")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

g = sns.catplot(data=net[net['Type'] == 'TV Show'], x='Decade', y='Duration', hue='Type', kind='bar')

g.set_axis_labels("Decade", "Duration")
g.set_titles("Relationship between TV Show (Seasons) Duration and Decade")
plt.tight_layout()
plt.show()
```
