4. Preparacao de variaveis

```python
airp = airp.drop(['Flight #','Registration', 'cn/In'], axis=1)
airp
```

```python
airp.rename(columns={'Operator':'Airp_comp','Type': 'Airp_model', 'Summary':'Crash_cause'}, inplace = True)
airp.head()
```

```python
import datetime as dt

airp['Date']=pd.to_datetime(airp['Date'])
airp['Year']=airp['Date'].dt.year
airp['Month']=airp['Date'].dt.month

airp.head()
```

```python
airp = airp.replace('c: 9:40', '09:40')
```

```python
airp['Time']=pd.to_datetime(airp['Time'], format='%H:%M')
airp['Hour'] = airp['Time'].dt.hour
airp.head()
```

```python
airp['Perc_survival'] = (1 - (airp['Fatalities']/airp['Aboard'])) * 100
airp['Perc_survival'] = round(airp['Perc_survival'],0)
airp.head()
```

```python
airp_gp_year_month = airp.groupby(['Year', 'Month'])
airp_gp_year_month.first()
```

```python
airp_gp_comp_model = airp.groupby(['Airp_comp','Airp_model'])
airp_gp_comp_model.first()
```

```python
airp_comp_fatal = airp.groupby('Airp_comp')['Fatalities'].sum()
airp_comp_fatal.sort_values()
```

```python
airp_small = airp[airp['Aboard'] <= 50]
airp_small.sort_values(by='Aboard').head()
```

```python
airp_medium = airp[(airp['Aboard'] > 50) & (airp['Aboard'] < 200)]
airp_medium.sort_values(by='Aboard').head()
```

```python
airp_large = airp[airp['Aboard'] >= 200]
airp_large.sort_values(by='Aboard').head()
```

```python
airp_ru_loc_surv50 = airp[(airp['Location'] == 'Moscow, Russia') & (airp['Perc_survival'] > 50)]
airp_ru_loc_surv50.head()
```

Visualizacao dos Dados

```python
airp_cl=pd.read_csv('cleaned_Airplane_Crashes_and_Fatalities_Since_1908.csv')
airp_cl.head()
```

```python
sns.pairplot(airp_cl)
```

```python
airp_cl['Decade'] = (airp_cl['Year'] // 10) * 10
plt.figure(figsize=(10, 6))

sns.countplot(data=airp_cl, x='Decade', palette='viridis')
plt.xlabel('Decade')
plt.ylabel('Number of Events')
plt.title('Number of Events per Decade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=airp_cl, x='Aboard', y='Fatalities', hue='Year', alpha=0.7)
plt.xlabel('Number of Aboard Passengers')
plt.ylabel('Number of Fatalities')
plt.title('Relationship between Aboard Passengers and Fatalities')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

location_counts = airp_cl['Location'].value_counts().head(10) 
location_counts.plot(kind='bar', color='skyblue')

plt.xlabel('Location')
plt.ylabel('Number of Events')
plt.title('Top 10 Locations with Most Events')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
combined_airp = pd.concat([airp_small.assign(Category='Small Plane'),
                           airp_medium.assign(Category='Medium Plane'),
                           airp_large.assign(Category='Large Plane')])

plt.figure(figsize=(12, 6))
dark_palette = sns.color_palette("dark:#2ecc71", n_colors=3)

sns.countplot(data=combined_airp, x='Category', palette=dark_palette)

plt.xlabel('Passenger Category')
plt.ylabel('Number of Events')
plt.title('Number of Events by Passenger Category')
plt.tight_layout()
plt.show()
```

```python
top_airp_models = airp_cl.groupby('Airp_model')['Fatalities'].sum().sort_values(ascending=False).head(10).index
top_airp_data = airp_cl[airp_cl['Airp_model'].isin(top_airp_models)]

plt.figure(figsize=(12, 6))
colorful_palette = sns.color_palette("Paired", n_colors=len(top_airp_models))

sns.boxplot(data=top_airp_data, x='Airp_model', y='Fatalities', palette=colorful_palette)
plt.xlabel('Airplane Model')
plt.ylabel('Number of Fatalities')
plt.title('Distribution of Fatalities for Top 10 Airplane Models')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
