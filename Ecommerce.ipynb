4. Preparacao de variaveis

```python
eco.rename(columns={'Avg. Session Length':'Avg_session','Time on App':'T_App', 'Time on Website':'T_Web', 'Length of Membership':'L_Member', 'Yearly Amount Spent':'T_Year_Spent'}, inplace=True)
eco.head()
```

```python
eco['Avg_session']=round(eco['Avg_session'],2)
eco['T_App']=round(eco['T_App'],2)
eco['T_Web']=round(eco['T_Web'],2)
eco['L_Member']=round(eco['L_Member'],0)
eco['T_Year_Spent']=round(eco['T_Year_Spent'],0)
eco.head()
```

```python
eco_app_av = eco.groupby(['T_App','L_Member'])
eco_app_av.first().sort_values(by='L_Member')
```

```python
eco.to_csv('cleaned_ecommerce_customers.csv')
```

Visualizacao dos dados

```python
sns.pairplot(eco)
plt.show()
```

```python
sns.jointplot(x = eco['T_Year_Spent'],y = eco['L_Member'], data = eco)
```

```python
sns.jointplot(x = eco['T_App'], y = eco['L_Member'], data = eco)
```

```python
sns.displot(eco['T_App'],kde=True)
```

```python
sns.displot(eco['T_Web'],kde=True)
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=eco, x='T_App', kde=True, label='T_App', color='blue')

sns.histplot(data=eco, x='T_Web', kde=True, label='T_Web', color='orange')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of T_App and T_Web')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 

top_avatar_counts = eco['Avatar'].value_counts().head(10)

top_avatar_counts.plot(kind='bar', color='skyblue')

plt.xlabel('Avatar')
plt.ylabel('Count')
plt.title('Top Ten Value Counts of Avatars')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```
