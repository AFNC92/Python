Bibliotecas em Python para Análise de Dados:
O Python começou como uma linguagem de programação de propósito geral. No entanto, a melhor legibilidade do Python tornou-o uma boa ferramenta para análise de dados.

## 2. Funcionalidades da biblioteca PLOTLY


O Plotly é uma biblioteca de visualização de dados em Python que oferece uma ampla variedade de funcionalidades para criar gráficos interativos e visualizações de dados de alta qualidade. Aqui está uma breve descrição das principais funcionalidades da biblioteca Plotly:

- **Gráficos Interativos:** permite criar gráficos interativos que podem ser explorados diretamente em um ambiente de notebook ou incorporados em sites e aplicativos da web. Os gráficos são interativos, o que significa que os usuários podem interagir com eles, fazer zoom, passar o cursor sobre pontos de dados para obter informações detalhadas e muito mais.

- **Suporte a Diversos Tipos de Gráficos:** oferece suporte a uma ampla variedade de tipos de gráficos, incluindo gráficos de dispersão, gráficos de barras, gráficos de linhas, gráficos Pie Chart, Heatmaps, gráficos 3D e muito mais. Isso torna a biblioteca versátil para atender às necessidades de diferentes tipos de dados e análises.

- **Personalização:** permite personalizar praticamente todos os aspectos dos gráficos, desde a aparência visual, como cores e estilos, até os detalhes do layout, como rótulos de eixo, títulos e legendas. Proporciona controlo total sobre a estética dos gráficos.

- **Integração com Pandas:** integra-se bem com estruturas de dados populares, como o Pandas, facilitando a criação de gráficos a partir de DataFrames e séries. Especialmente útil ao lidar com análises de dados tabulares.

- **Exportação e Partilha:** Os gráficos criados com o Plotly podem ser exportados para diferentes formatos, como imagens estáticas ou arquivos interativos. Além disso, é possível partilhar os gráficos interativos em páginas da web ou incorporá-los em aplicativos.

- **Suporte Multiplataforma:** não se limita ao Python. Também oferece suporte a outras linguagens de programação, como R, JavaScript, MATLAB e Julia, o que o torna uma escolha versátil para cientistas de dados e engenheiros de dados que trabalham em várias plataformas.

- **APIs para Personalização Avançada:** Para aqueles que desejam realizar personalizações avançadas ou incorporar gráficos em aplicativos da web personalizados, o Plotly oferece APIs avançadas para manipular os gráficos programaticamente.

Em resumo, o Plotly é uma biblioteca de visualização de dados poderosa e flexível que facilita a criação de gráficos interativos e personalizados para análises de dados e comunicação visual de informações. É amplamente utilizado em uma variedade de campos, incluindo ciência de dados, análise de negócios, pesquisa acadêmica e muito mais.


O Plotly possui três módulos principais. São eles:

1. plotly.plotly
2. plotly.graph_objects
3. plotly.tools

- plotly.plotly age como a interface entre a máquina local e o Plotly. Contém funções que requerem uma resposta do servidor do Plotly.

- plotly.graph_objects contém os objetos (Figura, layout, dados e a definição dos gráficos, como gráfico de dispersão e gráfico de linhas) que são responsáveis por criar os gráficos. A Figura pode ser representada como um dicionário (dict) ou como instâncias de plotly.graph_objects.Figure, e essas são serializadas como JSON antes de serem passadas para o plotly.js. Considere o exemplo abaixo para uma melhor compreensão.

Observação: o módulo plotly.express pode criar a Figura completa de uma só vez. Utiliza o módulo graph_objects internamente e retorna uma instância de graph_objects.Figure. Eis um exemplo:

```python
import plotly.express as px

# Creating the Figure instance
fig = px.line(x=[1,2, 3], y=[1, 2, 3])

# printing the figure instance
print(fig)
```

As figuras no Plotly são representadas como árvores, onde o nó raiz possui três atributos na camada superior: data, layout e frames, e os nós nomeados são chamados 'atributos'. No exemplo mencionado, layout.legend é um dicionário aninhado, onde "legend" é a chave dentro do dicionário, cujo valor também é um dicionário.

- plotly.tools contém diversas ferramentas na forma de funções que podem aprimorar a experiência com o Plotly. Essas ferramentas podem ser usadas para executar tarefas específicas relacionadas à criação, personalização e manipulação de gráficos e figuras no Plotly. São projetadas para simplificar o processo de criação de visualizações de dados interativas e personalizadas, tornando mais fácil para os desenvolvedores e cientistas de dados trabalharem com a biblioteca Plotly de maneira eficiente e conveniente.

Um gráfico de dispersão é um conjunto de pontos representando dados individuais nos eixos horizontal e vertical por meio de pontos. - ### Gráfico de setores (Pie charts):

Uma representação gráfica estatística circular, que é dividida em fatias para ilustrar proporções numéricas. Representa um tipo especial de gráfico que utiliza "fatias de pizza", onde cada setor mostra os tamanhos relativos dos dados. Um gráfico circular é dividido em segmentos, que descrevem frequências relativas ou magnitude e é conhecido como gráfico de círculo.

Podemos criar um gráfico de pizza usando a função go.Pie(). A função go.Pie() recebe os seguintes argumentos:
- labels: É uma lista de strings que especifica os rótulos das fatias.
- values: É uma lista de números que especifica os valores das fatias.
- name: É uma string que especifica o nome do gráfico de pizza.
- title: É uma string que especifica o título do gráfico de pizza.

```python
#Simple Pie Chart
marks = [48 , 30 , 20 , 15]
status = ['Excellent' , 'Good' , 'Average' , 'Bad']
data = go.Pie(
values= marks,
labels= status,
)
layout = go.Layout(
title=dict(text = "Marks by grade",x=0.46,y=0.95,font_size=20)
)
fig = go.Figure(data=data,layout=layout)
fig.show()
```

Os gráficos circulares são utilizados para compreender a composição dos dados e analisar as relações de parte para todo nos dados. Os gráficos circulares (e os gráficos de rosca) representam a composição percentual de um valor em comparação com o conjunto de dados/valor completo.

```python
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```

- ### Gráfico de Gantt;

Um gráfico GANTT (Gráfico de Atividades de Normalização Geral) é um tipo de gráfico no qual uma série de linhas horizontais é apresentada, mostrando a quantidade de trabalho realizado ou produção concluída num determinado período de tempo em relação à quantidade planeada para esses projetos. Esse tipo de gráfico é usado para representar cronogramas de projetos, permitindo que as equipas de projeto visualizem o progresso das tarefas ao longo do tempo e façam comparações com os planos originais. Cada linha horizontal representa uma atividade ou tarefa e a posição e o comprimento da linha indicam quando a atividade ocorre e quanto tempo leva para ser concluída. O gráfico GANTT é uma ferramenta valiosa para o gestao de projetos e o acompanhamento do progresso das tarefas.

```python
import plotly.figure_factory as ff 

df = [dict(Task="A", Start='2020-01-01', Finish='2009-02-02'),
    dict(Task="Job B", Start='2020-03-01', Finish='2020-11-11'),
    dict(Task="Job C", Start='2020-08-06', Finish='2020-09-21')]
 
# Creating the plot
fig = ff.create_gantt(df)
fig.show()
```

```python
df = [dict(Task="Job-1", Start='2017-01-01', Finish='2017-02-02', Resource='Complete'),
      dict(Task="Job-1", Start='2017-02-15', Finish='2017-03-15', Resource='Incomplete'),
      dict(Task="Job-2", Start='2017-01-17', Finish='2017-02-17', Resource='Not Started'),
      dict(Task="Job-2", Start='2017-01-17', Finish='2017-02-17', Resource='Complete'),
      dict(Task="Job-3", Start='2017-03-10', Finish='2017-03-20', Resource='Not Started'),
      dict(Task="Job-3", Start='2017-04-01', Finish='2017-04-20', Resource='Not Started'),
      dict(Task="Job-3", Start='2017-05-18', Finish='2017-06-18', Resource='Not Started'),
      dict(Task="Job-4", Start='2017-01-14', Finish='2017-03-14', Resource='Complete')]

colors = {'Not Started': 'rgb(220, 0, 0)',
          'Incomplete': (1, 0.9, 0.16),
          'Complete': 'rgb(0, 255, 100)'}

fig = ff.create_gantt(df, colors=colors, index_col='Resource', show_colorbar=True,
                      group_tasks=True)
fig.show()
```

- ### Gráfico Boxplot

Os gráficos de caixa (Box Plots) são uma ótima maneira de entender a distribuição de dados. Representam dados numéricos usando quartis, que são estatísticas que dividem os dados em quatro partes iguais. Um gráfico de caixa exibe a mediana (segundo quartil) como uma linha no meio de uma caixa retangular, que representa o intervalo interquartil (IQR), ou seja, a diferença entre o primeiro quartil (Q1) e o terceiro quartil (Q3). As "caudas" do gráfico de caixa se estendem a partir da caixa até os valores mínimos e máximos dos dados, geralmente com a inclusão de valores atípicos.

Os gráficos de caixa são úteis para identificar a dispersão e a presença de valores extremos nos dados, bem como para visualizar a simetria ou assimetria na distribuição dos dados. Fornecem uma representação visual compacta e informativa da distribuição dos dados, sendo valiosos em análises estatísticas e na compreensão da variabilidade dos dados numéricos.

```python
fig = px.box(stud, y="math score")
fig.show()
```

```python
fig = px.box(heart, x = 'slope',#the name of the column whose values are going to be used to place marks on the x-axis. Here, since we have slope as a categorical variable with 3 distict values, we will end up with a box plot for each slope
             y='chol',# here in this case, this is the column whose values be used to plot the box plot
             title = 'Distribution of Cholestrol across various slope', color = 'sex')
fig.show()
```

```python
fig = px.box(heart, x = 'slope',y='chol',title = 'Distribution of Cholestrol across various slope with Underlying data', 
             color = 'sex', 
             points = 'all' # we plot the underlying data as well of the plots
            )
fig.show()
```

- ### Violin Plot

Um método para visualizar a distribuição de dados numéricos de diferentes variáveis é através de um gráfico de violino (Violin Plot). Semelhante ao Box Plot, mas com um gráfico rotacionado em cada lado, fornecendo mais informações sobre a estimativa de densidade no eixo y. A densidade é espelhada e invertida, e a forma resultante é preenchida, criando uma imagem que se assemelha a um violino. A vantagem de um gráfico de violino é que ele pode mostrar nuances na distribuição que não são perceptíveis em um boxplot. Por outro lado, o boxplot mostra de forma mais clara os valores atípicos nos dados.

Em resumo, o gráfico de violino é uma ferramenta eficaz para visualizar a distribuição de dados numéricos, fornecendo informações sobre a forma da distribuição e a densidade dos dados, ao mesmo tempo em que destaca os valores atípicos. Especialmente útil quando se deseja obter uma compreensão mais detalhada da distribuição dos dados em diferentes variáveis.

```python
df = px.data.tips()
fig = px.violin(df, x="day", y="total_bill")
fig.show()
```

```python
fig = px.violin(
    heart, x = 'slope',y='chol',title = 'Distribution of Cholestrol across various slope with Underlying data', 
    color = 'sex', 
    points = 'all' #we plot the underlying data as well of the plots
               )
fig.show()
```

- ### Gráfico com mapas (Map Plot)

Os Map Plots no Plotly são uma forma poderosa de visualizar dados geográficos em forma de mapas interativos. Permitem representar dados em coordenadas geográficas, como latitude e longitude, de maneira eficaz, tornando mais fácil entender padrões e distribuições geográficas.

Aqui estão algumas das principais características e tipos de Map Plots oferecidos pelo Plotly:

1. **Scattergeo Plot (Gráfico de Dispersão Geográfica):** Este tipo de gráfico de mapa é semelhante a um gráfico de dispersão convencional, mas com os pontos de dados dispostos em locais geográficos específicos, usando coordenadas de latitude e longitude. É útil para exibir a distribuição geográfica de dados e é altamente interativo, permitindo que os utilizadores interajam com os pontos no mapa.

2. **Choropleth Map (Mapa de Coropletas):** Este tipo de gráfico de mapa é usado para representar dados geográficos por regiões, como países, estados ou municípios. Cada região é colorida de acordo com um valor numérico específico, permitindo uma rápida compreensão de tendências geográficas e variações de dados.

- ### Heatmap

Um heatmap é uma representação gráfica de dados que utiliza cores para visualizar os valores de uma matriz. Nesse tipo de gráfico, cores mais claras, geralmente tons de vermelho, são usadas para representar valores mais comuns ou atividades mais intensas, enquanto cores mais escuras são preferidas para representar valores menos comuns ou atividades menos intensas. O heatmap também é conhecido pelo nome de "matriz de sombreamento" devido à sua natureza de sombreamento da matriz de dados.

Os heatmaps são frequentemente utilizados para visualizar a distribuição e padrões em dados tabulares, onde as cores indicam a intensidade ou o valor dos dados em cada célula da matriz. São especialmente úteis quando se deseja identificar tendências, relações e áreas de maior ou menor concentração em grandes conjuntos de dados. Essa representação gráfica facilita a identificação de padrões visuais e a interpretação rápida dos dados. Os heatmaps são amplamente aplicados em várias áreas, como ciência de dados, biologia, finanças, análise de mercado e muito mais.

```python
data=[[1, 25, 30, 50, 1], [20, 1, 60, 80, 30], [30, 60, 1, 5, 20]]
fig = px.imshow(data,
                labels=dict(x="Day of Week", y="Time of Day", color="Productivity"),
                x=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
                y=['Morning', 'Afternoon', 'Evening']
               )
fig.update_xaxes(side="top")
fig.show()
```

- ### Gráficos 3D

Gráficos 3D usando a biblioteca Plotly em Python para visualizar dados tridimensionais de maneira interativa. 
Plotly oferece suporte a vários tipos de gráficos 3D, como gráficos de dispersão, superfícies, gráficos de barras e muito mais.

1. **3D Line Plots:**
O gráfico de linha no Plotly é uma adição muito acessível e ilustre ao Plotly, que permite gerir uma variedade de tipos de dados e montar estatísticas fáceis de estilizar. Com px.line_3d, cada posição de dados é representada como um vértice (cuja localização é dada pelas colunas x, y e z) de uma marca de polilinha no espaço 3D.

```python
# data to be plotted 
df = px.data.tips()  
  
# plotting the figure 
fig = px.line_3d(df, x="sex", y="day",  
                 z="time", color="sex")  
  
fig.show() 
```

2. **3D Scatter Plot:** Um gráfico de dispersão 3D pode representar gráficos bidimensionais que podem ser aprimorados mapeando até três variáveis adicionais, usando semântica de parâmetros de cor, tamanho e estilo. Todos estes parâmetros controlam a semântica visual, que é usada para identificar os diferentes subconjuntos de dados. O uso de semântica redundante pode ser útil para tornar os gráficos mais acessíveis. Pode ser criado usando a função scatter_3d da classe plotly.express.

```python
# Data to be plotted 
df = px.data.iris()  
  
# Plotting the figure 
fig = px.scatter_3d(df, x = 'sepal_width',  
                    y = 'sepal_length',  
                    z = 'petal_width',  
                    color = 'species')  
  
fig.show() 
```

3. **3D Surface Plot:**
Um gráfico de superfície (Surface Plot) é um tipo de gráfico tridimensional que representa dados com três variáveis: X, Y e Z. Em vez de mostrar pontos de dados individuais, o gráfico de superfície exibe uma relação funcional entre a variável dependente Y e duas variáveis independentes X e Z. Este tipo de gráfico é usado para distinguir entre variáveis dependentes e independentes num conjunto de dados.

O gráfico de superfície é particularmente útil quando se deseja visualizar como uma variável dependente (geralmente representada no eixo Z) varia em resposta a diferentes combinações de duas variáveis independentes (representadas nos eixos X e Y). Cria uma superfície tridimensional onde a altura ou cor da superfície representa o valor da variável dependente em relação às variáveis independentes.

Esses gráficos são frequentemente usados em campos como ciência, engenharia, matemática e análise de dados para visualizar relações complexas entre variáveis e identificar tendências ou padrões em dados tridimensionais. Eles são uma ferramenta valiosa para explorar a relação entre variáveis em um contexto tridimensional.

```python
# Data to be plotted 
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))  
y = x.copy().T  
z = np.cos(x ** 2 + y ** 2)  
  
# plotting the figure 
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])  
  
fig.show()
```

- ### Nuvem de palavras (World Cloud)

Uma "word cloud" é uma representação visual de dados de texto em que as palavras são exibidas em tamanhos variados com base na sua frequência ou importância num texto ou conjunto de dados específico. Numa "word cloud", as palavras que aparecem com maior frequência são geralmente exibidas com letras maiores, enquanto as palavras menos comuns são apresentadas com letras mais pequenas.

As "word clouds" são frequentemente usadas para visualização de dados e podem proporcionar uma forma rápida e intuitiva de identificar as palavras mais significativas ou frequentemente mencionadas num corpo de texto. São comumente utilizadas em áreas como análise de texto, análise de sentimento e exploração de dados.

Numa "word cloud", a disposição e arranjo das palavras são geralmente aleatórios ou seguem um padrão simples, como a ordem alfabética. A paleta de cores pode ser escolhida por motivos estéticos ou para transmitir informações adicionais, mas não representa valores de dados específicos.

Embora as "word clouds" sejam visualmente atrativas e possam proporcionar uma visão rápida das palavras mais proeminentes num texto, têm limitações. Não oferecem informações detalhadas sobre o contexto das palavras ou as relações entre elas. Além disso, podem não ser adequadas para dados textuais altamente estruturados ou complexos.

Em resumo, as "word clouds" são uma forma simples e acessível de visualizar a frequência das palavras em dados de texto, mas são mais bem utilizadas em conjunto com outras técnicas de análise de texto para uma compreensão mais abrangente da informação textual.

```python
from wordcloud import WordCloud  # Import WordCloud class from the wordcloud library

# data prepararion
x2011 = timesData.country[timesData.year == 2011]
plt.subplots(figsize=(8,8))
wordcloud = WordCloud(
                          background_color='white',
                          width=512,
                          height=384
                         ).generate(" ".join(x2011))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('graph.png')

plt.show()
```
